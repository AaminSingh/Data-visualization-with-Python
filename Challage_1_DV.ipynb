````python
books=pd.read_csv(r"C:\Users\vedas\OneDrive\ドキュメント\weather report01.csv")
books
````

````python
plt.plot(books["Date"],books["Temperature"])
plt.title('Daily high temperature')
plt.bar('Temperature','Date',color="purple")
plt.xlabel('Date')
plt.ylabel('Temperature')
````
